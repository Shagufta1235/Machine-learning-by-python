```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Sample dataset
emails = [
    ("Win a free iPhone now!", 1),
    ("Important meeting tomorrow", 0),
    ("Congratulations! You won a lottery!", 1),
    ("Project deadline extended", 0),
    ("Claim your free gift card", 1),
    ("Team meeting at 3 PM", 0),
    ("Urgent: Update your bank details", 1),
    ("Don't forget the dinner plan", 0)
]

# Convert to DataFrame
df = pd.DataFrame(emails, columns=['text', 'spam'])

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['spam'], test_size=0.2, random_state=42)

# Convert text to numerical data
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Train logistic regression model
model = LogisticRegression()
model.fit(X_train_vec, y_train)

# Make predictions
y_pred = model.predict(X_test_vec)

# Evaluate model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}\n")
print("Classification Report:\n", classification_report(y_test, y_pred))
```
